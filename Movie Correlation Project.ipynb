```python
# Import Libraries 
import pandas as pd
import numpy as np
import seaborn as sns

import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import matplotlib
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)

# Read in the data 
df = pd.read_csv(r'/Users/jess/Downloads/movies.csv')
original_df = pd.read_csv(r'/Users/jess/Downloads/movies.csv')
```

```python
# Let's look at the data 

df.head() 
```

```python
# Check for missing data 

for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, pct_missing))
```

```python
# Data Types for our columns 

df.dtypes
```

```python
# Changing data type of Budget and Gross Columns
    
df['budget'] = df['budget'].astype("Int64")
df['gross'] = df['gross'].astype("Int64")
```

```python
df.head()
```

```python
# Sort by highest gross to lowest gross

df = df.sort_values(by=['gross'], inplace=False, ascending=False)
original_df = df
```

```python
# display all the rows 

pd.set_option('display.max_rows', None)
```

```python
# Show all unique values

df['company'].drop_duplicates().sort_values(ascending=False).head()
```

```python
# Scatter plot with budget vs gross revenue

plt.scatter(x=df['budget'], y=df['gross'])
plt.title('Budget vs Gross Earning')
plt.xlabel('Gross Earning')
plt.ylabel('Budget for film')
plt.show()
```

```python
df.head()
```

```python
# Plot Budget vs Gross using seaborn 
# Adds a regression line

sns.regplot(x='budget', y='gross', data=df, scatter_kws={"color": "red"}, line_kws={"color": "blue"})
```

```python
# Finding correlation 

df.corr(numeric_only=True, method='pearson') #pearson, kendall, spearman
```

```python
# High correlation between budget and gross 

correlation_matrix = df.corr(numeric_only=True, method='pearson')
sns.heatmap(correlation_matrix, annot=True)
plt.title('Correlation Matric for Numeric Features')
plt.xlabel('Movie Features')
plt.ylabel('Budget for Film')
plt.show()
```

```python
# Looks at Company 

df.head()
```

```python
# Numerizing dataframe 

df_numerized = df 
for col_name in df_numerized.columns: 
    if(df_numerized[col_name].dtype == 'object'):
        df_numerized[col_name] = df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes
df_numerized.head()
```

```python
original_df.head()
```

```python
# Correlation for all fields

correlation_matrix = df_numerized.corr(numeric_only=True, method='pearson')
sns.heatmap(correlation_matrix, annot=True)
plt.title('Correlation Matric for Numeric Features')
plt.xlabel('Movie Features')
plt.ylabel('Budget for Film')
plt.show()
```

```python
df_numerized.corr()
```

```python
# Finding correlation for the whole df
correlation_mat = df_numerized.corr()

# Reshape df and pivot the head columns into head rows
corr_pairs = correlation_mat.unstack() 
corr_pairs
```

```python
# Sorting the pairs asc order

sorted_pairs = corr_pairs.sort_values()
sorted_pairs
```

```python
# Filtering only showing high correlation pairs

high_corr = sorted_pairs[(sorted_pairs) > 0.5]
high_corr.head()
```
